***Очереди***

В отличие от стека, элементы в очереди помещаются с одной стороны, а извлекаются с другой. Такой принцип управления данным называется **FIFO (First In First Out)**.

**<ins>Очередь на базе связного списка</ins>**

Новые эллементы мы так же добавляем в начало сразу после верхнего ограничителя. Вдобавок нам нужно отмечать первый элемент в очереди, например, с помощью атрибута first.

Чтобы при удалении первого элемента из списка не проходится по нему в цикле для смещения указателя на следующий элемент, для реализации очередей используется двунаправленный связный список. В таком случае при удалении элемента нужно будет только переопределить связи.

Методы добавления и удаления элементов в очереди называются **enqueue (добавить)** и **dequeue (удалить)**.

**<ins>Очередь на базе статического массива</ins>**

Для начала нужно ввести атрибут **last**, который будет указывать на последний добавленный элемент. Изначально равен -1.

При вставке элемента мы будем добавлять к last единицу, вычисляя новый индекс массива, а уже затем помещать в него данные.

При удалении элемента, если мы будем делать это в лоб, нам придется физически сдвигать все элементы влево, что занимает время O(N). Чтобы избежать лишних операций, в очередях на базе массивов добавляют дополнительный атрибут **first**, который ссылается на первый элемент в очереди. 

Теперь для извлечения данных нам достаточно вернуть значение элемента и сдвинуть атрибут first вправо. При этом в массиве по-прежнему останется такое же число элементов, а в очереди станет на 1 меньше.

Такое удаление происходит за время O(1), при этом происходит один интересный эффект: когда полностью заполнится массив, очередь будет заполнена только частично. В этом случае при добавлении нового элемента нам нужно закольцевать массив, передвинув указатель на последний элемент в начало (на позицию 0).

Таким образом, используя 2 указателя, мы не просто работаем за время O(1), но и в любой момент имеем доступ ко всем элементам массива.

**<ins>Очередь с приоритетом</ins>**

В очереди с приоритетом каждый элемент имеет свой вес или важность, и первым извлекатеся тот, что имеет наивысшийй приоритет.

*Пример*: такую структуру можно использовать для управления списком задач сотрудника: работник первыми выполняет задачи в порядке их приоритета, руководитель же наполняет список задач сотрудника с указанием их важности.

**Реализация очереди с приоритетом на базе связаного списка**

- Подход 1: сортированный двунаправленный связный список

Вставка элемента происходит за O(N), а извлечение - за O(1). 

- Подход 1: обычный двунаправленный связный список

Вставка элемента происходит за O(1), а извлечение - за O(N). 

**<ins>Дек</ins>**

Дек - это очередь с двухсторонним доступом, которая позволяет добавлять и удалять элементы с обеих сторон.

С помощью дека можно частично реализовать приоритетность, добавляя элементы с высоким приоритетом к одному концу очереди, а с низким - к другому.

Наиболее легко такие очереди реализовывать на базе двунаправленных связных списков. Мы же рассмотрим **реализацию дека на базе статического массива**.

Для этого необходимо запрограммировать 4 метода:

- push_back (синоним enqueue)

- push_front

- pop_back

- pop_front (синоним dequeue)
